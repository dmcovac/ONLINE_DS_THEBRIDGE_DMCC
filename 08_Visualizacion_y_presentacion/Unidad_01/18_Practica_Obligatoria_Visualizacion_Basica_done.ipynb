### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón,con trasparencia)




```python
import matplotlib.pyplot as plt

# Ordenamos por población y tomamos las 5 primeras
top_pobladas = df_cities.sort_values("population_total", ascending=False).head(5)

colores = ["blue", "gray", "olive", "brown", "lightcoral"]
plt.figure(figsize=(8,5))
bars = plt.bar(top_pobladas["city"], top_pobladas["population_total"], color=colores, alpha=0.7)

# Agregamos texto encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500, int(yval), ha="center")

plt.title("Ciudades más pobladas")
plt.ylabel("Población")
plt.show()
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
plt.figure(figsize=(8,5))
plt.hist(df_cities["area_total_km2"], bins=50, color="steelblue")

plt.title("Área total ocupada por las ciudades", fontsize=18, fontname="Bradley Hand ITC")
plt.xlabel("Área (Km2)")
plt.ylabel("Frecuencia")
plt.show()
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
plt.figure(figsize=(8,5))
plt.scatter(df_cities["area_water_km2"], df_cities["population_total"], alpha=0.5)

plt.title("Área de agua vs población")
plt.xlabel("Área de agua (km²)")
plt.ylabel("Población")
plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize=(5,5))
plt.boxplot(df_cities["elevation_m"].dropna())

plt.title("Boxplot de la altura de las ciudades")
plt.ylabel("Altura (m)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creamos figura y ejes
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Visualización de ciudades de California", fontsize=20)

# 1. Ciudades más pobladas
top_pobladas = df_cities.sort_values("population_total", ascending=False).head(5)
colores = ["blue", "gray", "olive", "brown", "lightcoral"]

bars = axs[0, 0].bar(top_pobladas["city"], top_pobladas["population_total"], color=colores, alpha=0.7)
for bar in bars:
    yval = bar.get_height()
    axs[0, 0].text(bar.get_x() + bar.get_width()/2, yval + 500, int(yval), ha="center")

axs[0, 0].set_title("Top 5 ciudades más pobladas")
axs[0, 0].set_ylabel("Población")

# 2. Histograma del área total 
axs[0, 1].hist(df_cities["area_total_km2"], bins=50, color="steelblue")
axs[0, 1].set_title("Área total ocupada (50 bins)", fontsize=18, fontname="Bradley Hand ITC")
axs[0, 1].set_xlabel("Área (km²)")
axs[0, 1].set_ylabel("Frecuencia")

# 3. Scatter: área agua vs población
axs[1, 0].scatter(df_cities["area_water_km2"], df_cities["population_total"], alpha=0.5)
axs[1, 0].set_title("Área de agua vs Población")
axs[1, 0].set_xlabel("Área de agua (km²)")
axs[1, 0].set_ylabel("Población")

# 4. Boxplot de altura de ciudades
axs[1, 1].boxplot(df_cities["elevation_m"].dropna())
axs[1, 1].set_title("Altura de las ciudades (Boxplot)")
axs[1, 1].set_ylabel("Altura (m)")

plt.tight_layout(rect=[0, 0, 1, 0.95])  # deja espacio al título general
plt.show()
```
